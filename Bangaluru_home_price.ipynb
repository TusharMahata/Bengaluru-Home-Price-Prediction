```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return (float(tokens[0]) + float(tokens[1]))/2

  try:
    return float(x)
  except:
    return None
```

```python
def remove_ppsf_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    sd = np.std(subdf.price_per_sqft)
    reduse_df = subdf[(subdf.price_per_sqft>(m-sd)) & (subdf.price_per_sqft<(m+sd))]
    df_out = pd.concat([df_out, reduse_df], ignore_index=True)
  return df_out

df7 = remove_ppsf_outliers(df6)
df6.shape
df7.shape
```

```python

from pickleshare import stat
def remove_bhk_outliers(df):
  exclusive_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std' : np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]

      }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclusive_indices = np.append(exclusive_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclusive_indices, axis='index')


df8 = remove_bhk_outliers(df7)
```

```python
df10 = df9.drop(['price_per_sqft'], axis='columns')
df10.head(30)
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head(5)
```

```python

df11 = pd.concat([df10, dummies.drop('other', axis = 'columns')], axis='columns')
df11.head(10)
```

```python
df11 = df11.drop('location', axis = 'columns')
```

```python
df11.shape
```

```python
X = df11.drop('price', axis = 'columns')
X.head(3)
```

```python
y = df11.price
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=16, test_size=.2, random_state=0)
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
def predict_price(location, sqft, bath, bhk, balcony):
  loc_index = np.where(X.columns==location)[0][0]

  x = np.zeros(len(X.columns))
  x[0] = sqft
  x[1] = bath
  x[2] = balcony
  x[3] = bhk
  if loc_index >= 0:
    x[loc_index] = 1
  return lr_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar', 5000, 7, 6, 6)
```
